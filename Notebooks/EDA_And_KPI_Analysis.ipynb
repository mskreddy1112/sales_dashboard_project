Performing Exploratory Data Analysis(EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

#Load the cleaned Data

```python
df = pd.read_csv("Cleaned_Superstore.csv", parse_dates=['Order Date', 'Ship Date'])
```

# set plotting style

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
```

# 1. Sales vs Region (Bar Chart)

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', color='skyblue')
plt.title("Total Sales by Region")
plt.ylabel("Sales")
plt.xlabel("Region")
plt.tight_layout()
plt.show()
```

#2. Profit vs Category & Sub-Category (Bar Chart)

```python
plt.figure(figsize=(12, 5)) # Adjust figure size as needed

# Plot Profit by Category
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
category_profit = df.groupby('Category')['Profit'].sum().sort_values()
category_profit.plot(kind='barh', color='coral')
plt.title("Profit by Category")
plt.ylabel("Profit")
plt.xlabel("Category")
plt.xticks(rotation=0) # Rotate x-axis labels if needed

# Plot Profit by Sub-Category
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()
subcat_profit.plot(kind='barh', color='lightgreen')
plt.title("Profit by Sub-Category")
plt.xlabel("Profit")
plt.ylabel("Sub-Category")
plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
```

# 3. Monthly Sales Trend (Line Chart)


```python
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales.index = monthly_sales.index.to_timestamp()

monthly_sales.plot(kind='line', marker='o', figsize=(12, 6))
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 4. Discount vs Profit (Scatterplot)

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category', alpha=0.7)
plt.title("Discount vs Profit")
plt.tight_layout()
plt.show()
```

# 5. Top Customers by Sales

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='bar', color='orchid')
plt.title("Top 10 Customers by Sales")
plt.ylabel("Sales")
plt.xlabel("Customer Name")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#Key Performance Indicators (KPIs)

```python
# Total Sales
total_sales = df['Sales'].sum()
print(f"Total Sales: â‚¹{total_sales:,.2f}")
```

```python
# Total Profit
total_profit = df['Profit'].sum()
print(f"Total Profit: â‚¹{total_profit:,.2f}")
```

```python
# Average Profit Margin
avg_margin = (df['Profit'].sum() / df['Sales'].sum()) * 100
print(f"Average Profit Margin: {avg_margin:.2f}%")
```

```python
# Top 5 Performing Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Products by Sales:")
print(top_products)
```

```python
# Monthly Sales Growth %
monthly_sales_df = df.groupby('Month')['Sales'].sum().to_frame()
monthly_sales_df['Growth %'] = monthly_sales_df['Sales'].pct_change() * 100
print("\nMonthly Sales Growth %:")
print(monthly_sales_df)
```

```python
# Profit by Region
profit_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print("\nProfit by Region:")
print(profit_region)
```

```python
# Profit by Segment
profit_segment = df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
print("\nProfit by Segment:")
print(profit_segment)
```

#ðŸ”Ž Generate Business Insights

```python
#ðŸ“Œ 1. Region with Highest Profit Margin
region_stats = df.groupby('Region')[['Sales', 'Profit']].sum()
region_stats['Profit Margin (%)'] = (region_stats['Profit'] / region_stats['Sales']) * 100
region_stats = region_stats.sort_values(by='Profit Margin (%)', ascending=False)
print(region_stats)
```

```python
#âœ… Insight: "South Region has the highest profit margin at {value:.2f}%. Consider expanding business operations here."#
```

```python
#ðŸ“Œ 2. Underperforming Category with High Discounts
cat_stats = df.groupby('Category')[['Discount', 'Profit']].mean()
cat_stats = cat_stats.sort_values(by='Profit')
print(cat_stats)
```

```python
#âœ… Insight: "Office Supplies category shows high average discount (~{value:.2f}%) but the lowest average profit. Consider reevaluating pricing strategy."
```

```python
#ðŸ“Œ 3. Quarter-wise Sales Trend (to show Q4 peak)
df['Quarter'] = df['Order Date'].dt.to_period("Q")
quarter_sales = df.groupby('Quarter')['Sales'].sum().sort_index()
quarter_sales.plot(kind='bar', title="Quarterly Sales Trend", ylabel="Sales")
```

```python
#âœ… Insight: "Q4 shows the highest sales volume. Focus marketing and stock buildup around Octoberâ€“December."
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

# Prepare data for Prophet
monthly = df.groupby('Order Date')['Sales'].sum().resample('ME').sum().reset_index()
monthly = monthly.rename(columns={"Order Date": "ds", "Sales": "y"})

# Initialize and train model
model = Prophet()
model.fit(monthly)

# Future prediction (next 3 months)
future = model.make_future_dataframe(periods=3, freq='ME')
forecast = model.predict(future)

# Plot forecast
fig1 = model.plot(forecast)
plt.title("ðŸ“ˆ Sales Forecast for Next 3 Months")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```
